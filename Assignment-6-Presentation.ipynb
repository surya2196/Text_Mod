```python
##Q5:Negative feedback are feedback with 'reviews_rating' of 1. Plot a Word Cloud and please also add the chart to the ppt.
##What is the most frequently occurring word in the negative feedback word cloud of “reviews_text”?
```

```python
!pip install wordcloud
from wordcloud import WordCloud
from nltk.probability import FreqDist
```

```python
qn5df_str = df[df["reviews_rating"]==1]["reviews_text"].to_string()

wordcloud = WordCloud(width = 1600, height = 800, max_font_size = 200,background_color ='white').generate(qn5df_str)

pos_words = []
for w1 in qn5df_str:
    pos_words.append(w1)
pos_words[:10]

qn5_txt = pd.Series(pos_words).str.cat(sep = ' ')


plot.figure(figsize=(20,10))
plot.imshow(wordcloud)
plot.axis("off")


fdist=FreqDist(qn5_txt)
fdist.most_common(30)
```
